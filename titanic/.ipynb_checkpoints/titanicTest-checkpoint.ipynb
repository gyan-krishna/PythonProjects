```python
#Gender analysis
Sex = dataframe['Sex']
Surv = dataframe['Survived']

ms = fs = 0
md = fd = 0
m = f = 0
for i in range(Sex.shape[0]):
    if(Sex[i] == "male"):
        m+=1
        if(Surv[i] == 1):
            ms += 1
        else:
            md += 1
    if(Sex[i] == "female"):
        if(Surv[i] == 1):
            fs += 1
        else:
            fd += 1

        f+=1
        
labels  = ["male surviviors", "female surviviors", 
           "male deceased", "female deceased"]
pi_data = [ms, fs,md, fd]
explode = [0.2, 0.2, 0.1, 0.1]

plt.pie(pi_data, explode, labels,autopct='%1.1f%%',shadow=True)
plt.show()
```
