```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Plotting x and y values

```python
x = np.array([0,6])
y = np.array([0, 250])

plt.plot(x, y, marker='o')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('plot')
plt.show()
```

```python
x = np.array(range(1, 200, 10))
y = np.array(np.log(x))

plt.plot(x, y, marker='*')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('plot')
plt.show()
```

```python
y = np.array(range(0, 10))**3
plt.plot(y, marker='X')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('plot')
```

```python
x = np.array(np.linspace(0, 5*np.pi, 60))
a = np.sin(x)

n = 8

b = np.sin(x+np.pi/n*1)*1
c = np.sin(x+np.pi/n*2)*1
d = np.sin(x+np.pi/n*3)*1
e = np.sin(x+np.pi/n*4)*1
f = np.sin(x+np.pi/n*5)*1
g = np.sin(x+np.pi/n*6)*1
h = np.sin(x+np.pi/n*7)*1
plt.rcParams['figure.figsize'] = [20,5]

plt.plot(x,a, 'o',  ms = 4, label = 'a', linestyle = '--', lw = 2, color = 'r')
plt.plot(x,b, 'o',  ms = 4, label = 'b', linestyle = '--', lw = 2, color = 'b')
plt.plot(x,c, 'o',  ms = 4, label = 'c', linestyle = '--', lw = 2, color = 'g')
plt.plot(x,d, 'o',  ms = 4, label = 'd', linestyle = '--', lw = 1, color = 'violet')
plt.plot(x,e, 'o',  ms = 4, label = 'e', linestyle = '--', lw = 1, color = 'dodgerblue')
plt.plot(x,f, 'o',  ms = 4, label = 'f', linestyle = '--', lw = 1, color = 'orange')
plt.plot(x,g, 'o',  ms = 4, label = 'e', linestyle = '--', lw = 1, color = 'crimson')
plt.plot(x,h, 'o',  ms = 4, label = 'f', linestyle = '--', lw = 1, color = 'DarkMagenta')

font1 = {'family':'serif', 'color':'blue', 'size':20}
font2 = {'family':'serif', 'color':'red', 'size':15}

plt.grid(color = 'gray', linestyle='--', lw = 0.5)
plt.legend()
plt.ylabel('amplitude',fontdict=font2)
plt.xlabel('time',fontdict=font2)
plt.title('Diwali sin wave', fontdict=font1, loc = 'center')
```
