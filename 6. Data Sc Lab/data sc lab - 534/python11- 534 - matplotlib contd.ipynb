```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Scatter plot

```python
x1 = np.random.randint(10, size=10)
y1 = np.random.randint(100, size=10)
plt.scatter(x1, y1, label = 'day 1')

x2 = np.random.randint(10, size=10)
y2 = np.random.randint(100, size=10)
plt.scatter(x2, y2, label = 'day 2')

plt.title('random scatter plot')
plt.xlabel('random x values')
plt.ylabel('random y values')
plt.legend()
```

#### Scatter plot with color bar

```python
x = np.random.randint(1000, size=500)
y = np.random.randint(1000, size=500)
colors = np.random.randint(1000, size=500)


plt.scatter(x, y, c = colors, cmap='plasma', s = 5)
plt.title('color map demo with random colors')
plt.xlabel('random x values')
plt.ylabel('random y values')
plt.colorbar()
plt.show()
```

#### Size of the scatter points and alpha (opacity )

```python
x = np.random.randint(1000, size=100)
y = np.random.randint(1000, size=100)
size = np.random.randint(800, size=100)
colors = np.random.randint(1000, size=100)

plt.scatter(x, y, c = colors, s = size, alpha = 0.6, cmap='cool')
plt.title('color map demo with random sizes')
plt.xlabel('random x values')
plt.ylabel('random y values')
plt.colorbar()
plt.show()
```

#### Bar graphs demo

```python
x = np.array(['a', 'b', 'c', 'd', 'e'])
y = np.random.randint(20, size=5)
print(x.dtype)
plt.bar(x, y, width = 0.8)
plt.title('bar graph demo with random values')
plt.xlabel('random x values')
plt.ylabel('random y values')
plt.show()
```

```python
x = np.array(['a', 'b', 'c', 'd', 'e'])
y = np.random.randint(20, size=5)
print(x.dtype)
plt.barh(x,y, color = 'tomato', height=0.4)
plt.title('horizontal bar graph demo with random values')
plt.xlabel('random x values')
plt.ylabel('random y values')
plt.show()
```

```python
x = np.random.randint(50, size=100)
plt.hist(x, width =3 )
plt.title('histogram demo with random values')
plt.xlabel('random x values')
plt.ylabel('frequency')
plt.show()
```

#### pi-chart

```python
x = np.random.randint(50, size=3)
l = ['C', 'C++', 'Java']
c = ['tomato', 'orange', 'indigo']
e = [0.1,0,0] # explode
plt.pie(x, labels= l, colors=c, explode=e, shadow = True)

plt.legend()
plt.title('pie chart')
plt.show()
```
