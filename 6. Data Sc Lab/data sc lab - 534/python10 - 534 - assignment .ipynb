```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('std.csv')
```

```python
df.describe()
```

```python
eng = df['english.grade']
sci = df['sciences.grade']
lng = df['language.grade']
mth = df['math.grade']
name = df['name']

n = len(eng)
std = []
for i in range(n):
    s = []
    s.append(name[i])
    s.append(eng[i])
    s.append(sci[i])
    s.append(lng[i])
    s.append(mth[i])
    std.append(s)
std
```

```python
plt.rcParams['figure.figsize'] = [20,50]
plt.suptitle('MARK SHEET')
for i in range(n):
    plt.subplot(n, 4, i+1)
    plt.title(std[i][0])
    plt.bar( ['eng', 'science', 'language', 'math'] , std[i][1:])
```
